<H2>FLIGHT LEVEL INFORMATION (flight_df)</H2>

flight_df is the main dataset. One row per flight

```python
flight_df
```

```python
print("\n[flight_df] Data Types and Null Values:")
flight_df.info()
```

```python
print("\n[flight_df] Data Distribution Summary:")
print(flight_df.describe(include = "all"))
```

**Insights from flight_df**  
- From total_seats:  
    - Aircraft sizes vary significantly, from small regional jets with 50 seats to large mainline aircraft with 364 seats.  
    - The average aircraft size is approximately 123 seats.  

- From ground_time (mins):  
    - The average scheduled ground time is 183.8 minutes, while the actual average is slightly higher at 189.9 minutes, suggesting ground operations take longer than planned on average.  
    - There are potential data quality issues: the minimum scheduled ground time is a negative value (-980 minutes), and the maximum for both scheduled and actual is extremely high (33,327 minutes or ~23 days), which are likely errors and require investigation.  

- From minimum_turn (mins):  
    - The required minimum turn times range from a quick 29 minutes to a lengthy 155 minutes.  
    - 75% of flights have a minimum required turn time of 56 minutes or less, highlighting that most turnarounds are scheduled to be under an hour.  

- From scheduled_arrival_station:  
    - There are 188 unique destinations served from ORD in this dataset.  
    - The most frequent destination is LGA (LaGuardia Airport), with 188 flights.  

- From fleet_type and carrier:  
    - The fleet is diverse, with 21 unique aircraft types.  
    - The most common aircraft is the CRJ-550, which operated 1,357 flights.  
    - More flights are operated by Mainline (4,502 flights) than by regional Express carriers.  

VISUALIZATION

```python
# Convert relevant columns to datetime objects
time_cols = [
    'scheduled_departure_datetime_local',
    'scheduled_arrival_datetime_local',
    'actual_departure_datetime_local',
    'actual_arrival_datetime_local'
]
for col in time_cols:
    flight_df[col] = pd.to_datetime(flight_df[col], errors='coerce')
```

```python
#departure delay: acutal-scheduled departure
#ground time pressure = scheduled ground time - min turn minutes
flight_df['departure_delay'] = (flight_df['actual_departure_datetime_local'] - flight_df['scheduled_departure_datetime_local']).dt.total_seconds() / 60
flight_df['ground_time_pressure'] = flight_df['scheduled_ground_time_minutes'] - flight_df['minimum_turn_minutes']
```

```python
# Set up the plot style
sns.set_style("whitegrid")
    
#Histogram of Departure Delay
plt.figure(figsize=(10, 6))
sns.histplot(flight_df['departure_delay'], bins=50, kde=True)
plt.title('Distribution of Departure Delays (in Minutes)', fontsize=16)
plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='--', label='On-Time Departure')
plt.legend()
plt.show()

#Histogram of Ground Time Pressure
plt.figure(figsize=(10, 6))
sns.histplot(flight_df['ground_time_pressure'], bins=40, kde=False)
plt.title('Distribution of Scheduled Ground Time Buffer', fontsize=16)
plt.xlabel('Buffer (Scheduled Ground Time - Minimum Turn Time)')
plt.ylabel('Number of Flights')
plt.axvline(0, color='red', linestyle='--', label='No Buffer')
plt.legend()
plt.show()
```

```python
#Bar Chart of Top 15 Destinations
plt.figure(figsize=(12, 7))
flight_df['scheduled_arrival_station_code'].value_counts().nlargest(15).plot(kind='bar', color='skyblue')
plt.title('Top 15 Most Frequent Destinations from ORD', fontsize=16)
plt.xlabel('Destination Airport')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Scatter Plot of Total Seats vs. Departure Delay
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_seats', y='departure_delay', data=flight_df, alpha=0.5)
plt.title('Departure Delay vs. Total Seats', fontsize=16)
plt.xlabel('Total Seats on Aircraft')
plt.ylabel('Departure Delay (Minutes)')
plt.show()
```

```python
#Box Plot of Delay by Fleet Type (Top 10)
top_fleets = flight_df['fleet_type'].value_counts().nlargest(10).index
plt.figure(figsize=(14, 8))
sns.boxplot(x='fleet_type', y='departure_delay', data=flight_df[flight_df['fleet_type'].isin(top_fleets)], order=top_fleets)
plt.title('Departure Delay by Aircraft Fleet Type (Top 10)', fontsize=16)
plt.xlabel('Fleet Type')
plt.ylabel('Departure Delay (Minutes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#Box Plot of Delay by Destination (Top 10)
top_destinations = flight_df['scheduled_arrival_station_code'].value_counts().nlargest(10).index
plt.figure(figsize=(14, 8))
sns.boxplot(x='scheduled_arrival_station_code', y='departure_delay', data=flight_df[flight_df['scheduled_arrival_station_code'].isin(top_destinations)], order=top_destinations)
plt.title('Departure Delay by Destination Airport (Top 10)', fontsize=16)
plt.xlabel('Destination Airport')
plt.ylabel('Departure Delay (Minutes)')
plt.show()
```

**Insights from bag_df**

- **From general counts & dates:**  
    - A total of 687,245 bags were processed over the 15-day period covered in the dataset.  
    - The busiest day for baggage processing was August 2nd, 2025, with 52,900 bags handled.  

- **From bag_tag_unique_number (Data Quality):**  
    - While there are 687,245 total bags, there are only 667,155 unique bag tag numbers. This suggests there are duplicate entries in the dataset.  
    - One bag tag (BAGTAG_167284) appears 9 times, which is a data anomaly that should be investigated.

```python
print("\n[pnr_remark_df] Data Distribution Summary:")
print(pnr_remark_df.describe(include='all'))
```

```python
# frequency of special service request
print("\n[pnr_remark_df] Frequency of Each Special Service Request:")
print(pnr_remark_df['special_service_request'].value_counts())
```

**Insights from pnr_remark_df**  
- **From Special Service Request Frequency:**  
    - Wheelchair requests are the overwhelming driver of special assistance. Combined, all three wheelchair types (Airport, Manual, Electric) account for 50,004 out of 51,698 total requests, which is approximately 97% of all special services.  
    - "Airport Wheelchair" is the most common need by a very large margin, representing 88.5% of all special requests.  
    - "Unaccompanied Minor" is the next significant operational task, with 1,706 distinct requests that require dedicated staff procedures.  

- **From General Data Distribution:**  
    - A total of 51,698 special service requests were made across the dataset.  
    - There are only 4 unique types of special service requests captured in this dataset, indicating a focus on mobility and unaccompanied minors.  
    - There are potential data quality issues or complex bookings, as one PNR (PNR_186497) is associated with 30 separate special service remarks.  

VISUALIZATIONS

```python
#Top 10 Special Service Requests
#It directly shows which passenger needs are most common and require the most operational planning.
plt.figure(figsize=(12, 8))
ssr_counts = pnr_remark_df['special_service_request'].value_counts().nlargest(10)
sns.barplot(x=ssr_counts.values, y=ssr_counts.index, palette='muted')
plt.title('Top 10 Most Common Special Service Requests (SSRs)', fontsize=16, weight='bold')
plt.xlabel('Number of Requests')
plt.ylabel('Type of Special Service Request')
plt.tight_layout()
plt.show()
```

```python
ssr_counts = pnr_remark_df['special_service_request'].value_counts()
```

```python
#Bar Chart for a clear, quantitative comparison of all service requests, immediately showing the massive scale of "Airport Wheelchair" needs compared to all other services.
plt.figure(figsize=(10, 7))
sns.barplot(x=ssr_counts.values, y=ssr_counts.index, palette='viridis')
plt.title('Absolute Count of Special Service Requests', fontsize=16, weight='bold')
plt.xlabel('Total Number of Requests')
plt.ylabel('Service Request Type')

for index, value in enumerate(ssr_counts.values):
    plt.text(value, index, f' {value}', va='center')
plt.tight_layout()
plt.show()
```

```python
#Pie Chart for Proportional Impact: highlights the main story that ~97% of all special service requests are for wheelchairs, which is a critical insight for resource planning and defining flight difficulty.
wheelchair_total = ssr_counts.loc[ssr_counts.index.str.contains("Wheelchair")].sum()
unaccompanied_minor_total = ssr_counts.get("Unaccompanied Minor", 0)

proportional_data = pd.Series(
    [wheelchair_total, unaccompanied_minor_total],
    index=['All Wheelchair Types', 'Unaccompanied Minor']
)

plt.figure(figsize=(8, 8))
plt.pie(
    proportional_data,
    labels=proportional_data.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=['violet', 'lightgreen'],
    textprops={'fontsize': 12, 'weight': 'bold'}
)
plt.title('The Dominance of Wheelchair-Related Services', fontsize=16, weight='bold')
plt.ylabel('')
plt.show()
```
